````python
# 🧪 Backend API Performance Testing
import requests
import time
import json

BACKEND_URL = "http://localhost:8000"

def test_endpoint_performance(endpoint, method="GET", data=None):
    """Test API endpoint performance"""
    url = f"{BACKEND_URL}{endpoint}"
    headers = {'Content-Type': 'application/json'} if data else {}
    
    start_time = time.time()
    
    try:
        if method == "POST":
            response = requests.post(url, json=data, headers=headers, timeout=5)
        else:
            response = requests.get(url, timeout=5)
        
        response_time = (time.time() - start_time) * 1000  # Convert to ms
        
        return {
            'endpoint': endpoint,
            'status': response.status_code,
            'response_time_ms': round(response_time, 2),
            'success': response.status_code == 200,
            'data_size': len(response.content) if response.content else 0
        }
    except Exception as e:
        return {
            'endpoint': endpoint,
            'status': 'ERROR',
            'response_time_ms': (time.time() - start_time) * 1000,
            'success': False,
            'error': str(e)
        }

# Test critical endpoints
endpoints_to_test = [
    {
        'endpoint': '/api/health',
        'method': 'GET',
        'description': 'System health monitoring'
    },
    {
        'endpoint': '/api/planetary/current-hour',
        'method': 'POST',
        'data': {'location': {'lat': 37.7749, 'lon': -122.4194}},
        'description': 'Planetary hour calculations'
    },
    {
        'endpoint': '/api/tokens/calculate',
        'method': 'POST',
        'data': {
            'tokens': {'Spirit': 1.0, 'Essence': 0.8, 'Matter': 0.6, 'Substance': 0.4},
            'location': {'lat': 37.7749, 'lon': -122.4194}
        },
        'description': 'Token rate calculations'
    }
]

print("🧪 Backend API Performance Testing:")
print("====================================")

results = []
for test in endpoints_to_test:
    result = test_endpoint_performance(
        test['endpoint'], 
        test.get('method', 'GET'), 
        test.get('data')
    )
    results.append(result)
    
    status_icon = "✅" if result['success'] else "❌"
    print(f"{status_icon} {test['description']}")
    print(f"   Response Time: {result['response_time_ms']}ms")
    print(f"   Status: {result['status']}")
    print()

# Performance summary
successful_tests = [r for r in results if r['success']]
if successful_tests:
    avg_response_time = sum(r['response_time_ms'] for r in successful_tests) / len(successful_tests)
    print(f"📊 Average Response Time: {avg_response_time:.2f}ms")
    print(f"🎯 Performance Target (<100ms): {'✅ PASSED' if avg_response_time < 100 else '❌ NEEDS OPTIMIZATION'}")
````

````python
# 💰 Token Rate Calculation with Planetary Influences
import math
from datetime import datetime

def calculate_token_rates_with_planetary_influence(base_rates, planetary_hour, timestamp):
    """Calculate token rates influenced by planetary conditions"""
    
    # Planetary multipliers for each token type
    planetary_multipliers = {
        'Sun': {'Spirit': 1.3, 'Essence': 1.1, 'Matter': 0.9, 'Substance': 0.8},
        'Moon': {'Spirit': 0.9, 'Essence': 1.4, 'Matter': 1.1, 'Substance': 0.9},
        'Mercury': {'Spirit': 1.2, 'Essence': 0.9, 'Matter': 0.8, 'Substance': 1.3},
        'Venus': {'Spirit': 0.8, 'Essence': 1.3, 'Matter': 1.2, 'Substance': 1.0},
        'Mars': {'Spirit': 1.4, 'Essence': 0.8, 'Matter': 1.3, 'Substance': 0.7},
        'Jupiter': {'Spirit': 1.2, 'Essence': 1.2, 'Matter': 0.9, 'Substance': 1.1},
        'Saturn': {'Spirit': 0.7, 'Essence': 0.9, 'Matter': 1.4, 'Substance': 1.2}
    }
    
    # Apply planetary influences
    multipliers = planetary_multipliers.get(planetary_hour, {})
    planetary_rates = {
        token: base_rates[token] * multipliers.get(token, 1.0)
        for token in base_rates
    }
    
    # Apply temporal influences
    hour = timestamp.hour
    day_of_year = timestamp.timetuple().tm_yday
    
    # Time of day influence (dawn/dusk peaks)
    time_multiplier = 1 + 0.2 * math.sin((hour * math.pi) / 12)
    
    # Seasonal influence
    seasonal_multiplier = 1 + 0.15 * math.sin((day_of_year * 2 * math.pi) / 365)
    
    # Apply temporal modifiers
    final_rates = {
        token: rate * time_multiplier * seasonal_multiplier
        for token, rate in planetary_rates.items()
    }
    
    return {
        'rates': final_rates,
        'influences': {
            'planetary_hour': planetary_hour,
            'time_multiplier': time_multiplier,
            'seasonal_multiplier': seasonal_multiplier
        }
    }

# Example calculation
base_rates = {'Spirit': 1.0, 'Essence': 0.8, 'Matter': 0.6, 'Substance': 0.4}
current_time = datetime.now()
planetary_hour = 'Jupiter'

token_data = calculate_token_rates_with_planetary_influence(base_rates, planetary_hour, current_time)

print("💰 Token Rate Calculation Example:")
print(f"Time: {current_time.strftime('%Y-%m-%d %H:%M:%S')}")
print(f"Planetary Hour: {planetary_hour}")
print("\nBase Rates → Final Rates:")
for token in base_rates:
    base = base_rates[token]
    final = token_data['rates'][token]
    change = ((final - base) / base) * 100
    print(f"  {token}: {base:.3f} → {final:.3f} ({change:+.1f}%)")

print("\nInfluence Factors:")
for factor, value in token_data['influences'].items():
    if isinstance(value, (int, float)):
        print(f"  {factor}: {value:.3f}")
    else:
        print(f"  {factor}: {value}")
````
